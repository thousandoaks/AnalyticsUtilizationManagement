There have been a lot of studies undertaken in the past on factors affecting life expectancy considering demographic variables, income composition and mortality rates among others. The impact of immunization and human development index has not been taken into account though.

# About the Dataset




The Global Health Observatory (GHO) data repository under World Health Organization (WHO) keeps track of the health status as well as many other related factors for all countries The data-sets are made available to public for the purpose of health data analysis. 

The data-set related to life expectancy, health factors for 193 countries has been collected from the same WHO data repository website and its corresponding economic data was collected from United Nation website. Among all categories of health-related factors only those critical factors were chosen which are more representative. 

It has been observed that in the past 15 years , there has been a huge development in health sector resulting in improvement of human mortality rates especially in the developing nations in comparison to the past 30 years. 

This dataset considers data from year 2000-2015 for 179 countries for further analysis. The individual data files have been merged together into a single data-set.  The final merged file(final dataset) consists of 21 Columns and 2864 rows. All predicting variables was then divided into several broad categories:â€‹Immunization related factors, Mortality factors, Economical factors and Social factors.


## Factors:

Thinness_five_nine_years:

Average years that people aged 25+ spent in formal education

Economy_status_Developed:

### Q2. Fit a statistical model regression Life Expectancy in terms of QUANTITATIVE medical and economic factors
#### Tip. start with strongly correlated factors. Keep adding one factor at a time

```python
reg1 = smf.ols(formula='Life_expectancy~ Infant_deaths+Adult_mortality', data=LifeExpectancyDataFrame
)
results1=reg1.fit()
print(results1.summary())
```

```python
reg2 = smf.ols(formula='Life_expectancy~ Infant_deaths+Adult_mortality+Alcohol_consumption', data=LifeExpectancyDataFrame)

results2=reg2.fit()
print(results2.summary())
```

```python
LifeExpectancyDataFrame.columns.values
```

```python
reg3 = smf.ols(formula='Life_expectancy~ Infant_deaths+Adult_mortality+Alcohol_consumption+Diphtheria+Polio+Schooling', data=LifeExpectancyDataFrame)

results3=reg3.fit()
print(results3.summary())
```

```python
reg4 = smf.ols(formula='Life_expectancy~ Infant_deaths+Adult_mortality+Alcohol_consumption+Measles+Diphtheria+Polio+BMI+Schooling+GDP_per_capita', data=LifeExpectancyDataFrame)

results4=reg4.fit()
print(results4.summary())
```

#### Based on the previous table we notice that all factors considered in the model are statistically significant except for Measles,Diphteria and Polio (p-values larger than 0.05). We remove these 3 factors in subsequent models

```python
reg5 = smf.ols(formula='Life_expectancy~ Infant_deaths+Adult_mortality+Alcohol_consumption+BMI+Schooling+GDP_per_capita', data=LifeExpectancyDataFrame)

results5=reg5.fit()
print(results5.summary())
```


#### `Infant_deaths`, `Adult_mortality` and `BMI` DECREASE `Life_expectancy`.
#### `Alcohol_consumption`, `Schooling` and `GDP_per_capita` INCREASE `Life_expectancy`

#### Schooling seem to have a strong effect in life expectancy. One year increse in schooling increases life expectancy by 0.15 years.


#### We are finding a surprising effect: Alcohol seems to increase life span. This is probably due to the fact that wealthier societies drink more alcohol. This is the case given the strong positive levels of correlation between alcohol consumption and both GDP and developed status.

### Q3. Does economic development increase life expectancy ?
#### Tip. Consider adding the qualitative factor: `Economy_status_Developed`

```python
reg6 = smf.ols(formula='Life_expectancy~ Infant_deaths+Adult_mortality+Alcohol_consumption+BMI+Schooling+GDP_per_capita+Economy_status_Developed', data=LifeExpectancyDataFrame)

results6=reg6.fit()
print(results6.summary())
```

#### From the previous table we observe that the factor `Economy_status_Developed` is statistically significant (p-value=0). Being in a developed economy increases life expectancy by 0.51 years.

```python
sns.violinplot(data=LifeExpectancyDataFrame, x="Economy_status_Developed", y="Life_expectancy")
```

### Q3. Does the regional factor influence life expectancy ?

```python
reg7 = smf.ols(formula='Life_expectancy~ Infant_deaths+Adult_mortality+BMI+Schooling+GDP_per_capita+Economy_status_Developed+Region', data=LifeExpectancyDataFrame)

results7=reg7.fit()
print(results7.summary())
```

#### From the previous we conclude that there exists a regional effect as far as life expectancy is concerned

```python
sns.violinplot(data=LifeExpectancyDataFrame, x="Region", y="Life_expectancy")
plt.xticks(rotation=90);
```
